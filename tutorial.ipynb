# Reversible first-order reactions

For a reversible reaction $A \rightleftarrows B$ the rate law is of the form

$$
v = - \frac{d[A]}{dt} = k_f [A] - k_b [B]
$$

where $k_f$ is the rate constant of the forward reaction and $k_b$ is the rate constant of the backward reaction.

If both A and B are present initially, $[A]_0 = [A]$ and $[B]_0 = [B]$ then $[B] = [A]_0 + [B]_0 - [A]$ is a boundary condition:

$$
- \frac{d[A]}{dt} = (k_f + k_b) \left( [A] - \frac{k_b}{k_f + k_b} ([A]_0 + [B]_0) \right)
$$

Integrating and isolating $[A]$:

$$
[A] = -\frac{(k_b [B]_0 - k_f [A]_0) \exp(-t(k_f + k_b)) - k_b([A]_0 + [B]_0 }{k_f + k_b}
$$

The reactions of the mutual conversion of isomers, for example $d$-menthone $\rightleftarrows$ $l$-menthone or ammonium thiocyanate $\rightleftarrows$ thiocarbamide can serve as examples of reversible reactions.

## First example

To instantiate a class, pass $k_f$, $k_b$, $[A]_0$ and $[B]_0$:

```python
example01 = ReversibleFirstOrder(0.5, 1, 1, 0)  # kf, kb, [A]0, [B]0
```

Plotting Concentration vs Time:

```python
example01.plot_reaction_time()
```

Plotting Reaction quotient vs Time:

```python
example01.plot_reaction_quotient_time()
```

Please observe that the vertical axis uses scientific notation.

The kinetic attributes can be retrieved:

```python
print(f'[A]0 = {example01.A0}')
print(f'[B]0 = {example01.B0}')
print(f'kf = {example01.kf}')
print(f'kb = {example01.kb}')
```

Some plot parameters can be customized:

```python
example01.plot_reaction_time(time_unit='min', concentration_unit='mmol/L', 
                             title_text=r'Concentration vs time for A $\rightleftarrows$ B',
                             label_reactant='A', label_product='B')
```

```python
example01.plot_reaction_quotient_time(time_unit='min', q='Reaction quotient',
                                     title_text=r'Reaction quotient vs time for A $\rightleftarrows$ B')
```

The title text will not be displayed if the parameter `title` is set to `False`:

```python
example01.plot_reaction_time(time_unit='min', concentration_unit='mmol/L', 
                             title_text=r'Concentration vs time for A $\rightleftarrows$ B',
                             label_reactant='A', label_product='B', title=False)
```

```python
example01.plot_reaction_quotient_time(time_unit='min', q='Reaction quotient',
                                     title_text=r'Reaction quotient vs time for A $\rightleftarrows$ B',
                                     title=False)
```

## Interactive plot

The following must be imported for interactive plots:

```python
from ipywidgets import interact, interactive, fixed
import matplotlib.pyplot as plt
```

The plot methods from `ReversibleFirstOrder` return Matplotlib axes, so it is possible to create a figure with subplots. Since $k_f$, $k_b$, $[A]_0$ and $[B]_0$, as well as time, are instance attributes, we need to create a `ReversibleFirstOrder` instance for each plot:

```python
def plot_interactive(kf=1, kb=1, A0=1, B0=0, time=5):
    instance = ReversibleFirstOrder(kf, kb, A0, B0, time)
    fig, arr = plt.subplots(nrows=1, ncols=2, figsize=(14,6), constrained_layout=True, facecolor=(1,1,1))
    instance.plot_reaction_time(ax=arr[0], title_text=r'Concentration vs time for A $\rightleftarrows$ B',
                                label_reactant='A', label_product='B')
    instance.plot_reaction_quotient_time(ax=arr[1], title_text=r'Reaction quotient vs time for A $\rightleftarrows$ B',)
    plt.show()
```

Now, with `inteactive` we can change the attributes. Each change will create a new instance of `ReversibleFirstOrder`:

```python
graph = interactive(plot_interactive,
                    time=fixed(5),
                    kf=(1,10,1),
                    kb=(1,10,1),
                    A0=(0,10,1),
                    B0=(0,10,1))
display(graph)
```
